## Exploratory data analysis on sklearn.load_wine()

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
data = datasets.load_wine()
print(data)
```

```python
data.feature_names
```

```python
data.DESCR
```

```python
wine_data = pd.DataFrame(data['data'])
wine_data.columns = data.feature_names
wine_data.head()
```

```python
wine_data['target'] = data.target
wine_data.head()
```

##### Check for any null values 

```python
wine_data.isnull().values.any()
```

Let;s check the effect of amount to alcohol in wine on our target column.

```python
sns.catplot(x = 'target', y = 'alcohol', kind = 'swarm', data = wine_data)
```

#### We can see that all the target classes have different ranges of amount of alcohols.<br>
##### target_0 : 13.0 - 14.5<br>
##### target_1 : 11.5 - 13.0<br>
##### target_2 : 12.0 - 14.0<br>

```python
sns.catplot(x = 'target', y = 'proline', kind = 'box', data = wine_data)
```

#### Proline also show categorized wines easily as we can identify different ranges.
